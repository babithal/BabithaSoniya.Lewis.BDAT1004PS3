# Question 6
Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
var = pd.read_csv("us-marriages-divorces-1867-2014.csv")
```

```python
fig,x1 = plt.subplots()
x1.plot(var['Year'], var['Marriages_per_1000'], color="red", marker="o")
x1.set_xlabel("YEAR",fontsize=10, color ="green")
x1.set_ylabel("MARRIAGES",color="pink",fontsize=12)

fig,x2= plt.subplots()

x2.plot(var ['Year'], var['Divorces_per_1000'],color="black",marker="o")
x2.set_ylabel("DIVORCES",color="yellow",fontsize=12)
x2.set_xlabel("YEAR",fontsize=10, color ="blue")
plt.figure(figsize=(800,800))
plt.show()
```

# Question 7
Create a vertical bar chart comparing the number of marriages and divorces per
capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python
import numpy as np 
import matplotlib.pyplot as plt 
var = pd.read_csv("us-marriages-divorces-1867-2014.csv")
comp = var [(var.Year == 1900) | (var.Year == 1950) | (var.Year == 2000)]
Men =  comp ['Divorces_per_1000']
Women = comp ['Marriages_per_1000']
year = comp['Year']
year_axis = np.arange(len(year))
plt.bar(year_axis - 0.4, Women, 0.4, label = 'Women')
plt.bar(year_axis + 0.4, Men, 0.4, label = 'Men')
plt.xticks(year_axis, year)
plt.xlabel("Gender")
plt.ylabel("No of marriages and divorces")
plt.title("Number of marriages and divorces per capita in US")
plt.legend()
plt.show()
```

# Question 9
Create a pie chart showing the fraction of all Roman Emperors that were
assassinated.
Make sure that the pie chart is an even circle, labels the categories, and shows the
percentage breakdown of the categories.

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
death= pd.read_csv("roman_emperor.csv")
death=death[(death.Cause_of_Death == 'Assassinated')]
count = death['Length_of_Reign']
fig=plt.figure(figsize =(8,10))
plt.pie(count, labels =death['Emperor'],autopct='%1.0f%%')
plt.show()
```
